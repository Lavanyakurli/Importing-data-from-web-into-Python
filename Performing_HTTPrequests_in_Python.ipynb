```python
from urllib.request import urlopen, Request
```

```python
url="http://www.datacamp.com/teach/documentation"
```

```python
#GETRequest
req=Request(url)
```

```python
#packaging the request and storing the response of the GET Request
response=urlopen(req)
```

```python
print(type(response))
```

```python
#Reading the html content 
print(response.read())
response.close()
```

All the above steps can be performed in a single piece of code using requests package

```python
import requests 
```

```python
#package the request, send the request and collect the response
response=requests.get(url)
```

```python
#extracting the response
text=response.text
print(text)
```

# Tag Soup is a pejorative term that refers to syntactically or structurally incorrect HTML written for a web page.To make the html/XML look in a structured way BeautifulSoup package is used.
# Documentation of BeautifulSoup https://www.crummy.com/software/BeautifulSoup/bs4/doc/

```python
from bs4 import BeautifulSoup
```

```python
url2='https://www.python.org/~guido/'
```

```python
req=requests.get(url2)
```

```python
html=req.text
print(html)
#This html is unstructured with different tags
```

```python
#A beautifulsoup object is created using BeautifulSoup function
soup=BeautifulSoup(html)
```

```python
#Prettify() method structures the html for display
pretty_soup=soup.prettify()
```

```python
print(pretty_soup)
```

```python
#To print the title of the HTML, use title attribute on beautifulSoup object.
Title=soup.title
print(Title)
# title is one of the tag of any HTML
```

```python
#Extracting the text from HTML
Text=soup.text #or soup.get_text() can also be used
print(Text)
```

```python
#To get all the available hyperlinks in the HTML. Hyperlinks are defined by HTML tag <a>
tags=soup.find_all('a')
```

```python
print(type(tags))
#the variable tags is a result set. to print all the hyper links we have to enumerate over it using for loop and print the actual URLS
```

```python
for link in tags:
    print(link.get('href'))
```
